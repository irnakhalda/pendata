### b. Eksplorasi Data

Pada kode ini, dilakukan serangkaian tahapan eksplorasi awal terhadap dataset "ObesityDataSet_raw_and_data_sinthetic.csv" menggunakan pustaka pandas, seaborn, dan matplotlib. Tahap pertama adalah memuat data ke dalam DataFrame df menggunakan pd.read_csv(). Setelah data berhasil dimuat, langkah selanjutnya adalah melakukan pemeriksaan nilai hilang (missing values) pada setiap kolom dengan menggunakan fungsi isnull().sum(), yang akan memberikan jumlah nilai kosong pada tiap atribut. Pemeriksaan ini penting dilakukan untuk mengetahui apakah dataset memiliki data yang tidak lengkap, karena keberadaan missing value dapat memengaruhi proses analisis dan akurasi model machine learning.

Langkah selanjutnya adalah mendeteksi keberadaan outlier pada fitur-fitur numerik menggunakan metode IQR (Interquartile Range). Untuk setiap kolom numerik, dihitung nilai kuartil pertama (Q1) dan kuartil ketiga (Q3), kemudian dihitung rentang interkuartil (IQR = Q3 - Q1). Nilai dianggap sebagai outlier jika berada di bawah Q1 - 1.5 * IQR atau di atas Q3 + 1.5 * IQR. Jumlah outlier untuk setiap fitur kemudian dicetak. Tahapan terakhir adalah menampilkan distribusi dari kelas target, yaitu kolom NObeyesdad, yang menunjukkan tingkat obesitas dari setiap individu. Distribusi ini ditampilkan dalam bentuk jumlah kemunculan masing-masing kelas, dan sangat penting untuk mengetahui apakah data target bersifat seimbang atau tidak (class imbalance), karena hal ini akan memengaruhi pemilihan metode evaluasi dan strategi pelatihan model.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ----------------------------
# 0. LOAD DATA
# ----------------------------
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')

# ----------------------------
# 1. CEK MISSING VALUE
# ----------------------------
print("Missing Values :")
print(df.isnull().sum())
print()

# ----------------------------
# 2. CEK OUTLIER DENGAN IQR
# ----------------------------
print("Outlier Check (IQR Method):")
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outlier(s)")

print()

# ----------------------------
# 3. DISTRIBUSI KELAS TARGET
# ----------------------------
print("=== Distribusi Kelas Target ===")
target_col = 'NObeyesdad'
print(df[target_col].value_counts())
print()

```

## Prepocessing

Pada kode ini, dilakukan proses pra-pemrosesan data untuk menyiapkan dataset "ObesityDataSet_raw_and_data_sinthetic.csv" sebelum digunakan dalam pelatihan model machine learning. Pertama-tama, data dimuat menggunakan pandas.read_csv(), kemudian dipisahkan menjadi dua bagian utama: fitur (X) yang mencakup seluruh kolom kecuali NObeyesdad, dan target (y) yaitu kolom NObeyesdad yang merepresentasikan kategori tingkat obesitas dari masing-masing individu. Selanjutnya, kolom-kolom numerik dalam fitur dideteksi menggunakan select_dtypes(), dan dilakukan proses normalisasi menggunakan MinMaxScaler untuk mengubah rentang nilai menjadi antara 0 dan 1. Hal ini penting untuk menyamakan skala antar fitur, khususnya jika akan digunakan pada model-model yang sensitif terhadap skala seperti KNN atau SVM.

Setelah normalisasi data numerik, dilakukan encoding terhadap fitur kategorikal menggunakan LabelEncoder, agar semua data dapat diubah menjadi bentuk numerik yang dapat diproses oleh algoritma pembelajaran mesin. Proses encoding dilakukan dengan mengonversi setiap nilai kategorikal menjadi bilangan bulat unik. Seluruh data hasil encoding kemudian dibagi menjadi data latih (X_train, y_train) dan data uji (X_test, y_test) menggunakan train_test_split dengan rasio 80:20. Pada bagian akhir, kode mencetak informasi dimensi data hasil split dan contoh baris dari X_train untuk memastikan bahwa proses pra-pemrosesan telah berhasil dan data siap digunakan untuk proses klasifikasi atau pelatihan model lainnya.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# 1. Load dataset obesitas dari file lokal
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')

# 2. Pisahkan fitur (X) dan target (y)
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']

# 3. Deteksi kolom numerik
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# 4. Normalisasi data numerik
scaler = MinMaxScaler()
X_scaled = X.copy()
X_scaled[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# 5. Encoding kolom kategorikal
X_encoded = X_scaled.copy()
categorical_cols = X_encoded.select_dtypes(include='object').columns

le = LabelEncoder()
for col in categorical_cols:
    X_encoded[col] = le.fit_transform(X_encoded[col])

# 6. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# 7. Tampilkan informasi akhir
print("Bentuk data X_train:", X_train.shape)
print("Bentuk data X_test :", X_test.shape)
print("Contoh data X_train:")
print(X_train.head())
```

## Klasifikasi pada data
Kode ini digunakan untuk melakukan klasifikasi tingkat obesitas berdasarkan kebiasaan makan dan kondisi fisik, menggunakan tiga algoritma pembelajaran mesin: Decision Tree, Random Forest, dan Support Vector Machine (SVM). Dataset yang digunakan diambil secara langsung dari UCI Machine Learning Repository melalui pustaka ucimlrepo dengan ID 544, yang memuat data estimasi tingkat obesitas. Data yang telah diambil kemudian digabungkan antara fitur (X) dan target (y) ke dalam satu DataFrame df untuk keperluan encoding. Nama kolom target diidentifikasi secara otomatis dari objek target. Seluruh kolom bertipe kategorikal pada dataset kemudian diubah ke format numerik menggunakan teknik Label Encoding dengan bantuan LabelEncoder, agar data dapat diterima oleh algoritma machine learning.

Setelah proses encoding, fitur dan target dipisahkan kembali ke dalam X_encoded dan y_encoded, kemudian dilakukan pembagian data menjadi data latih dan data uji menggunakan train_test_split dengan proporsi 80% data latih dan 20% data uji. Sebelum model dilatih, dilakukan proses standarisasi menggunakan StandardScaler agar semua fitur numerik berada dalam skala yang seragam. Tiga model klasifikasi kemudian dilatih secara terpisah: Decision Tree, Random Forest, dan SVM. Setiap model diuji terhadap data uji dan dievaluasi menggunakan metrik akurasi serta classification report (precision, recall, dan f1-score untuk setiap kelas obesitas). Hasil dari evaluasi ini ditampilkan di konsol, sehingga memungkinkan pengguna untuk membandingkan performa masing-masing model dalam mengklasifikasikan tingkat obesitas berdasarkan data yang tersedia.

```python
# 1. Import Library
from ucimlrepo import fetch_ucirepo
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# 2. Fetch Dataset
data = fetch_ucirepo(id=544)  # Estimation of Obesity Levels
X = data.data.features.copy()
y = data.data.targets.copy()

# 3. Gabungkan fitur dan target untuk encoding
df = pd.concat([X, y], axis=1)

# 4. Ambil nama kolom target (otomatis)
target_column = y.columns[0]

# 5. Label Encoding untuk semua kolom kategorikal
categorical_cols = df.select_dtypes(include='object').columns
df_encoded = df.copy()
le_dict = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col])
    le_dict[col] = le

# 6. Pisahkan fitur dan target
X_encoded = df_encoded.drop(columns=[target_column])
y_encoded = df_encoded[target_column]

# 7. Split Data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# 8. Standarisasi
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ====================
# 9. TRAINING MODEL
# ====================

# a. Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_scaled, y_train)
dt_pred = dt.predict(X_test_scaled)

# b. Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_scaled, y_train)
rf_pred = rf.predict(X_test_scaled)

# c. SVM
svm = SVC(random_state=42)
svm.fit(X_train_scaled, y_train)
svm_pred = svm.predict(X_test_scaled)

print("=== Decision Tree ===")
print("Accuracy:", accuracy_score(y_test, dt_pred))
print(classification_report(y_test, dt_pred))

print("=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))

print("=== SVM ===")
print("Accuracy:", accuracy_score(y_test, svm_pred))
print(classification_report(y_test, svm_pred))
```

```python
# 1. Import Library
from ucimlrepo import fetch_ucirepo
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings("ignore")  # menghilangkan warning dari xgboost

# 2. Fetch Dataset
data = fetch_ucirepo(id=544)
X = data.data.features.copy()
y = data.data.targets.copy()

# 3. Gabungkan fitur dan target
df = pd.concat([X, y], axis=1)

# 4. Ambil nama kolom target (otomatis)
target_column = y.columns[0]

# 5. Label Encoding
categorical_cols = df.select_dtypes(include='object').columns
df_encoded = df.copy()
le_dict = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col])
    le_dict[col] = le

# 6. Pisahkan fitur dan target
X_encoded = df_encoded.drop(columns=[target_column])
y_encoded = df_encoded[target_column]

# 7. Split Data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# 8. Standarisasi
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ====================
# 9. TRAINING MODEL
# ====================
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "XGBoost": XGBClassifier(eval_metric='mlogloss', random_state=42)
}

model_scores = {}

# ====================
# 10. EVALUASI + CONFUSION MATRIX
# ====================
for name, model in models.items():
    print(f"\n=== {name} ===")
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)

    # Akurasi dan klasifikasi
    acc = accuracy_score(y_test, y_pred)
    model_scores[name] = acc
    print(f"Akurasi: {acc:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=le_dict[target_column].classes_,
                yticklabels=le_dict[target_column].classes_)
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

# ====================
# 11. KESIMPULAN MODEL TERBAIK
# ====================
best_model = max(model_scores, key=model_scores.get)
print("=== KESIMPULAN ===")
print("Akurasi masing-masing model:")
for model, score in model_scores.items():
    print(f"{model}: {score:.4f}")

print(f"\n Model dengan akurasi terbaik adalah: **{best_model}** dengan akurasi {model_scores[best_model]:.4f}")

# ====================
# 12. VISUALISASI PERBANDINGAN AKURASI
# ====================
plt.figure(figsize=(8, 5))
sns.barplot(x=list(model_scores.keys()), y=list(model_scores.values()), palette="viridis")
plt.title("Perbandingan Akurasi Tiap Model")
plt.ylabel("Akurasi")
plt.ylim(0, 1)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
