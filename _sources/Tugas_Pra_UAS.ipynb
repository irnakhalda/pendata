```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ----------------------------
# 0. LOAD DATA
# ----------------------------
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')

# ----------------------------
# 1. CEK MISSING VALUE
# ----------------------------
print("Missing Values :")
print(df.isnull().sum())
print()

# ----------------------------
# 2. CEK OUTLIER DENGAN IQR
# ----------------------------
print("Outlier Check (IQR Method):")
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outlier(s)")

print()

# ----------------------------
# 3. DISTRIBUSI KELAS TARGET
# ----------------------------
print("=== Distribusi Kelas Target ===")
target_col = 'NObeyesdad'
print(df[target_col].value_counts())
print()

```

## Visualisasi Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Load data
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')

# Ambil kolom numerik
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Loop setiap kolom numerik
for col in numerical_cols:
    fig = plt.figure(figsize=(10, 5))

    # Subplot histogram (atas)
    ax1 = plt.subplot2grid((2, 1), (0, 0), rowspan=1)
    sns.histplot(data=df, x=col, kde=True, bins=30, color='skyblue', ax=ax1)
    ax1.set_title(f'Histogram: {col}')
    ax1.set_xlabel('')

    # Subplot boxplot (bawah)
    ax2 = plt.subplot2grid((2, 1), (1, 0), rowspan=1)
    sns.boxplot(data=df, x=col, color='orange', ax=ax2)
    ax2.set_title(f'Boxplot: {col}')

    plt.tight_layout()
    plt.show()
```

## Prepocessing

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# 1. Load dataset obesitas dari file lokal
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')

# 2. Pisahkan fitur (X) dan target (y)
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']

# 3. Deteksi kolom numerik
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# 4. Normalisasi data numerik
scaler = MinMaxScaler()
X_scaled = X.copy()
X_scaled[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# 5. Encoding kolom kategorikal
X_encoded = X_scaled.copy()
categorical_cols = X_encoded.select_dtypes(include='object').columns

le = LabelEncoder()
for col in categorical_cols:
    X_encoded[col] = le.fit_transform(X_encoded[col])

# 6. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# 7. Tampilkan informasi akhir
print("Bentuk data X_train:", X_train.shape)
print("Bentuk data X_test :", X_test.shape)
print("Contoh data X_train:")
print(X_train.head())
```
